```python
import numpy as np
import pandas as pd
from numpy import linalg as LA
from scipy.signal import find_peaks
from matplotlib import pyplot as plt
```

```python
splits = []
bnch1 = pd.read_csv("Data_serie_test/extracted_bench_serie1.csv", index_col=0, parse_dates=True)
bnch2 = pd.read_csv("Data_serie_test/extracted_bench_serie2.csv", index_col=0, parse_dates=True)
sqt = pd.read_csv("Data_serie_test/extracted_squat_serie.csv", index_col=0, parse_dates=True)
bnch1
```

```python
dt = bnch2[['X_arm','Y_arm','Z_arm','X_feet','Y_feet','Z_feet']]
dt = np.array(dt[:-35])


dt = LA.norm(dt,  axis=1)

peaks, _ = find_peaks(dt, distance=20)
plt.plot(dt)
plt.plot(peaks, dt[peaks], "x")
plt.plot(np.zeros_like(dt), "--", color="gray")
plt.show()

a = np.convolve(dt,[1,1,1], 'same')

peaks, _ = find_peaks(a, distance=20)
plt.plot(a)
plt.plot(peaks, a[peaks], "x")
plt.plot(np.zeros_like(a), "--", color="gray")
plt.show()

a = np.convolve(a,[-1,0,1], 'same')


peaks, _ = find_peaks(a, distance=20, height=10)
plt.plot(a)
plt.plot(peaks, a[peaks], "x")
plt.plot(np.zeros_like(a), "--", color="gray")
plt.show()
```
